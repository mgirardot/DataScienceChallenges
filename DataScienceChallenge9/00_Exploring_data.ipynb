```python
import pandas as pd
import numpy as np
```

```python
filePath = "/home/michael/Documents/syslog-v3.3/20010411.log"
```

```python
with open(filePath) as f:
    data = f.read().split('\n')
```

```python
data[:3]
```

```python
elements = []
for line in data:
    el = line.split()
    try:
        elements.append([el[0],el[4], ' '.join(el[7:])])
    except:
        print el
```

```python
elements[:10]
```

```python
df = pd.DataFrame(elements, columns=['timestamp','AP','message'], dtype=(np.datetime64,np.str,np.str))
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.head()
```

```python
df['Disassociating'] = df.message.str.extract(r'Disassociating (\w+)', expand=False)
df['Station'] = df.message.str.extract(r'Station (\w+)', expand=False)
```

```python
df.head()
```

```python
df['Action'] = df.message.str.extract(r'Station \w+ (\w+)', expand=False)
```
